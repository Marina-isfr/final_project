# message_if

```python
#my_highlights

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_features=40)

# Fit and transform the 'about_me' column
X = vectorizer.fit_transform(df['message_if'])

# Convert the result to a DataFrame
sekeywords_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
sekeywords_df.info()

# love, fun, people, meet, enjoy, humor, friends, meet, talk, love

#typical_friday  47221 non-null  object 
# 30  my_secret       47221 non-null  object 
# 31  message_if 
```

```python
# after analizing the 10 columns im creating 20 binary columns that summarize 
# relevant aspects for each profile
import pandas as pd
import re

# Manually defined keywords
keywords = ['love', 'fun', 'cooking', 'music', 'read', 'friends', 'work', 'travel', 'humor', 'movie', 'laugh', 'new', 'personality','future','people', 'smile','talk','life','tv','play']
# Columns to check
columns_to_check = ['about_me', 'my_goals', 'my_talent', 'my_highlights', 'my_favorites' , 'my_needs', 'think_about','typical_friday', 'my_secret', 'message_if']  

# Function to create binary columns
def keyword_presence(text, keyword):
    pattern = re.compile(r'\b' + re.escape(keyword) + r'\w*\b', re.IGNORECASE)
    return 1 if pattern.search(text) else 0

# Create binary columns for each keyword
for keyword in keywords:
    df[keyword] = df[columns_to_check].apply(lambda row: any(keyword_presence(str(cell), keyword) for cell in row), axis=1)

# Display the resulting DataFrame
print(df)

#travel, family, friends, fun, passionate, laugh, work, love, enjoy, new
#life, work, school, work, new, people, job, friends, people, time
#life, people, love, cooking, friends, laugh, playing, fun, dancing, listening
# hair, humor, eyes, laugh, smile, tall, personality, people, pretty, height
# books, music, movie, shows, tv, eat, food, read, life, family
# books, music, internet, coffee, friends , laughter, phone, family, love, food
# friends, future, work, love, people, time, travel, world, music, family
# movie, work, bar, dancing, dinner, friends, hanging, music, relaxing, playing
# open, think, private, love, people, admit, ask , tell, love, want
# love, fun, people, meet, enjoy, humor, friends, meet, talk, love

```
