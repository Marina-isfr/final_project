```python
# Initialize the OneHotEncoder
ohe = OneHotEncoder()

# Fit and transform the categorical columns
ohe_encoded = ohe.fit_transform(dfhab[['diet', 'drinks', 'drugs', 'smokes']])

# Create a DataFrame with the one-hot encoded columns
ohe_df = pd.DataFrame(ohe_encoded.toarray(), columns=ohe.get_feature_names_out(['diet', 'drinks', 'drugs', 'smokes']), index=dfhab.index)

# Concatenate with the original DataFrame
df_habits_numerical = pd.concat([dfhab[['profile_id']], ohe_df], axis=1)

# Print the new DataFrame
print(df_habits_numerical)
```
