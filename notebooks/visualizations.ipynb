# Column age & Height per sex

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Fill missing values with mean
df['age'] = df['age'].fillna(df['age'].mean())
df['height'] = df['height'].fillna(df['height'].mean())

# Convert height to centimeters if needed
if df['height'].max() < 3:  # assuming heights are in meters
    df['height'] = df['height'] * 100

# Define the palette with lighter colors
palette = sns.color_palette(["#a6dba0", "#c2a5cf"])  # Lighter green and purple

# Set plot style
sns.set_style("white")

# Columns to plot
num_cols = ['age', 'height']
xlims = [(15, 70), (130, 220)]  # Age limits and height limits in cm
titles = ['Age Distribution', 'Height Distribution']

# Create subplots
fig2, ax2 = plt.subplots(ncols=2, figsize=(14, 5), sharey=False)

# Plot each histogram
for i, col in enumerate(num_cols):
    sns.histplot(data=df, x=col, ax=ax2[i], hue='sex', element='step', bins=40, alpha=0.4, palette=palette)
    ax2[i].set(xlim=xlims[i], title=titles[i])

# Set specific colors for 'sex' categories
for patch, label in zip(ax2[0].patches, df['sex'].unique()):
    if label == 'male':
        patch.set_facecolor(palette[0])  # Light green for male
    elif label == 'female':
        patch.set_facecolor(palette[1])  # Light purple for female

plt.tight_layout()
plt.show()
```
